```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas dataframe.

city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head()
```

```python
# Read the ride data file and store it in a pandas dataframe.

ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset

pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_df.head(10)
```

```python
# Create the Urban city DataFrame.

urban_cities_df = pyber_data_df[pyber_data_df["type"]=="Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban and Rural city DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"]=="Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"]=="Rural"]

rural_cities_df.head()
```

```python
# Get the number of rides for urban cities. Use groupby() function to create a Series of data 
# that has the name of the city as the index
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
# Create the suburban and rural ride count. Use groupby() again.
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]

rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
# Get average fare for each city in the urban cities.
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
# Get average fare for each city in the suburban and rural cities.
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
```

```python
# Get the average number of drivers for each urban city.
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
# Get the average number of drivers for each city for the suburban and rural cities.
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots (bubble chart) for urban cities, and add labels, title, legend and other formats.
plt.scatter(urban_ride_count, 
            urban_avg_fare, 
           s=10*urban_driver_count,
           c="coral",
           edgecolor="k",
           lw=1,
           alpha=0.8, label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for suburban cities.
plt.scatter(suburban_ride_count, 
            suburban_avg_fare, 
           s=10*suburban_driver_count,
           c="skyblue",
           edgecolor="k",
           lw=1,
           alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter plots for rural cities.
plt.scatter(rural_ride_count, 
            rural_avg_fare, 
           s=10*rural_driver_count,
           c="gold",
           edgecolor="k",
           lw=1,
           alpha=0.8, label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Add the scatter charts for each type of city.
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="k", lw=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="k", lw=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="k", lw=1,
      alpha=0.8, label="Rural")

# Show the plot
plt.show()
```

```python
# Edit the existing (upper) code.
# Build the scatter charts for each city type.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="k", lw=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="k", lw=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="k", lw=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties (new edits)
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Add the legend.
plt.legend()
# Show the plot
plt.show()
```

```python
# Create a legend. Declare a variable for the legend function, lgnd = plt.legend(), 
# and add parameters for font size, legend location, and legend title, along with some other features.
# Edit the existing (upper) code.
# Build the scatter charts for each city type.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="k", lw=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="k", lw=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="k", lw=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties (new edits)
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

# Create the legend.
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

plt.show()

# Description of the "lgnd" code

# 1. We made the font size for the text "small" with fontsize=.
# 2. Then we expanded the legend horizontally using mode= to fit the area. 
#.    Because the font size is small, this is optional.
# 3. We added the number of scatter points in the legend for each marker to be 1. 
#.    We can add multiple marker points by increasing the number.
# 4. The location setting, loc=, for the legend is where it will fit the "best" based on 
#.    the plotting of the data points.
# 5. We added a legend title.
# 6. We set each marker in the legend equal to 75-point font by using the legendHandles[]._sizes 
#     attribute and list indexing in the brackets to reference one of the three scatter plots. 
# 7. We increased the font size of the legend title to 12.

# Save the figure.
plt.savefig("analysis/Fig1.png")
```

```python
# Get the sum of the fares for each city type. Using groupby() function on the pyber_data_df datagrame
# group by the type of city of column so the type of city is the index. Sum up the fare column and divide fare
# for each type of city and select fare column.

sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
# Get the sum of all fares.
total_fares = pyber_data_df["fare"].sum()
total_fares
```

```python
# Calculate the percentage of fare for each city type.
type_percents = 100 * sum_fares_by_type / total_fares
type_percents
```

```python
# We can do the code in one line too.
type_percents = 100 * pyber_data_df.groupby(["type"]).sum()["fare"] / pyber_data_df["fare"].sum()
type_percents
```

```python
# Build the percentage of fares by city type pie chart.

plt.pie(type_percents, labels=["Rural", "Suburban", "Urban"])
plt.show()
```

```python
# Format the pie. 

plt.pie(type_percents,
    labels=["Rural", "Suburban", "Urban"], colors=["gold", "skyblue", "coral"], 
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
    
plt.title("Percentage of Total Fares by City Type")
plt.show()
```

```python
# Import mpl to change the plot configurations using rcParams. 
import matplotlib as mpl

# Build Pie Chart
plt.subplots(figsize=(10, 60))

plt.pie(type_percents,
    labels=["Rural", "Suburban", "Urban"], colors=["gold", "skyblue", "coral"], 
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
    
plt.title("Percentage of Total Fares by City Type")

# Change the default size from 10 to 14. For this we need to import matplotlib which is different from
# matplotlib.pyplot and use rcParams to change the pie chart's parameters.

mpl.rcParams["font.size"] = 14

plt.savefig("analysis/Fig5.png")

plt.show()
```
